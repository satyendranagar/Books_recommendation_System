# Page 1

```python
ratings_with_name = pd.merge(ratings, books, on="ISBN")
print(ratings_with_name.shape)
ratings_with_name.head()
```

```python
temprary_df = ratings_with_name.groupby("Book-Title").agg({"Book-Rating" : ['count', 'mean']}).reset_index()
popular_df = temprary_df["Book-Rating"].rename(columns = {"count" : "num_ratings", "mean" : "avg_ratings"})
popular_df["Book-Title"] = temprary_df["Book-Title"]
popular_df.head()
```

```python
popular_df = popular_df[popular_df["num_ratings"] > 250].sort_values("avg_ratings", ascending=False)[0:50]
popular_df.head()
```

```python
popular_df = popular_df.merge(books, on="Book-Title").drop_duplicates("Book-Title")[['Book-Title', 'Book-Author',
       'Year-Of-Publication', 'Publisher', 'Image-URL-L', 'num_ratings', 'avg_ratings']]
popular_df.head(7)
```

```python
pickle.dump(popular_df, open("Popular_books.pkl", "wb"))
a = popular_df["Image-URL-L"].values
a[2]
```

# Page 2

```python
temporary = ratings_with_name.groupby("User-ID").count()["Book-Rating"] > 200
critics = temporary[temporary].index
user_filtered_rating = ratings_with_name[ratings_with_name["User-ID"].isin(critics)]
print(user_filtered_rating.shape)
user_filtered_rating.head()
```

```python
temporary = user_filtered_rating.groupby("Book-Title").count()["Book-Rating"] >= 50
popular_books = temporary[temporary].index
final_rating = user_filtered_rating[user_filtered_rating["Book-Title"].isin(popular_books)]
print(final_rating.shape)
final_rating.head()
```

```python
pt = final_rating.pivot_table(index="Book-Title", columns="User-ID", values="Book-Rating", fill_value=0)
pt.head()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity_score = cosine_similarity(pt)
similarity_score.shape
```

```python
def recommend(book_name):
    index = np.where(pt.index==book_name)[0][0]
    similar_books = sorted(list(enumerate(similarity_score[index])),key = lambda x:x[1], reverse = True)[1:6]
    
    for i in similar_books:
        print(pt.index[i[0]])
```

```python
recommend("The Notebook")
```
